```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Yangi data
data = {
    "kv_m": [35, 45, 55, 65, 75, 85, 95],
    "narxi": [18000, 25000, 32000, 42000, 53000, 65000, 78000]
}

df = pd.DataFrame(data)

print("üè† Yangi Uy Ma'lumotlari:")
print(df)

print("\nüìä DataFrame haqida umumiy ma'lumot:")
print(df.info())

print("\nüìà Statistik tahlil:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["kv_m"], df["narxi"], marker='o', color='green')
plt.title("Uy maydoni va narxi o'rtasidagi bog'liqlik")
plt.xlabel("Kvadrat metr")
plt.ylabel("Narxi (ming so'm)")
plt.grid(True)
plt.show()
```
