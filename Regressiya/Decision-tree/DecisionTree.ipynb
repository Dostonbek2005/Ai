```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

# Yangi ma'lumotlar
X_new = np.array([[50], [65], [78], [84], [90], [98]])
y_new = np.array([13000, 18000, 23000, 41000, 48000, 60000])

# Yangi model
model_new = DecisionTreeRegressor(max_depth=2)
model_new.fit(X_new, y_new)

# Daraxtni chizish
plt.figure(figsize=(10, 6))
plot_tree(model_new, filled=True, feature_names=["House Size"], rounded=True)
plt.title("New Decision Tree")
plt.show()
```
